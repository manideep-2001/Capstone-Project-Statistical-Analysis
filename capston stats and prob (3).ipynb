```python
# maximum confirmed cases
India.confirmed.max()
```

```python
# day by day the conformed cases are increased 
India.plot(x="date", y=["confirmed" ], kind="line", figsize=(10, 9))
```

```python
# we can clearly observe that the peopel fully vaccinated are very high compared to conformed cases. 
#this means the vaccination process was way faster than the increasing conformed cases
India.plot(x="date", y=["confirmed","people_fully_vaccinated" ], kind="line", figsize=(10, 9))
```

```python
# as we observe in graph we can clearnly see that the transportation was effected a lot compared to working and school
# school was the least effected 
India.plot(x="date", y=["school_closing","workplace_closing","transport_closing" ], kind="hist", figsize=(10, 9))
```

```python
# as we observed from graphs we can clearly observe that international moment was taken more serious then the stay home and internal movement and gathering restrictions
# As time passed, international movement restrictions became more serious.
India.plot(x="date", y=["gatherings_restrictions","stay_home_restrictions","internal_movement_restrictions","international_movement_restrictions" ], kind="hist", figsize=(10, 9))
```

```python
# facial covering was increased and remain constant at some point. 
# there was a huge increase of facial coverings comapred before covid 
# There was a sudden shift in facial coverings from not required to very required.
India.plot(x="date", y=["facial_coverings"], kind="line", figsize=(10, 9))
```
